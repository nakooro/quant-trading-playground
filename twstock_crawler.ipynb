```
import requests
import pandas as pd
import io
```

```
def crawler(yymmdd):
  url = "http://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date={}&type=ALLBUT0999".format(yymmdd)
  req = requests.get(url)
  csvList = req.text.splitlines()
  iniIndex = -1
  for i, text in enumerate(csvList):
    if text == '"證券代號","證券名稱","成交股數","成交筆數","成交金額","開盤價","最高價","最低價","收盤價","漲跌(+/-)","漲跌價差","最後揭示買價","最後揭示買量","最後揭示賣價","最後揭示賣量","本益比",':
      iniIndex = i
      break
  df = pd.read_csv(io.StringIO(''.join([text[:-1].replace('=','') + '\n' for text in csvList[iniIndex:]])))
  return df
```

```
import datetime
import time

def parse_date(yyyymmdd):
  return ''.join(str(yyyymmdd).split(' ')[0].split('-'))

def crawl_between(iniYYMMDD, endYYMMDD):
  iniDate = datetime.strptime(iniYYMMDD, '%Y%m%d').strftime('%m/%d/%Y')
  endDate = datetime.strptime(endYYMMDD, '%Y%m%d').strftime('%m/%d/%Y')

def crawl_n_days(iniDate, n):
  dfs = {}
  cur = iniDate
  for i in range(n):
    cur -= datetime.timedelta(days=1)
    parseDate = parse_date(cur)    
    try:
      df = crawler(parseDate)
      print('added {}'.format(parseDate))
      print(df.head())
      dfs.update({parseDate: df})
    except:
      print('Failed at {}'.format(parseDate))
    time.sleep(10)
    
  return dfs
```

```
# stocks = crawl_n_days(datetime.datetime.now(), 360 )
targetDate = datetime.date(2019,5,11)
stocks = crawl_n_days(targetDate, int(''+str((targetDate-datetime.date(2008,1,1)).days)))
```
